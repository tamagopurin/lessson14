```python
import numpy as np
import skfuzzy as fuzz
import matplotlib.pyplot as plt
import math
%matplotlib inline
```

```python
# 3.1
x = np.arange(-2, 2, .2)
y = np.arange(-2, 2, .2)
X, Y = np.meshgrid(x, y)
mur=math.e**(-1*(X**2+Y**2));

plt.rcParams['font.size'] = 14
plt.rcParams['font.family'] = 'Times New Roman'

# グラフの入れ物を用意する。
fig = plt.figure()
ax1 = fig.add_subplot(111, projection='3d')

# 軸のラベルを設定する。
ax1.set_xlabel('x')
ax1.set_ylabel('y')
ax1.set_zlabel('Membership to the Fuzzy Set R')

# データプロットする。
ax1.plot_surface(X, Y, mur, cmap='jet')

# グラフを表示する。
plt.show()
plt.close()
```

```python
# 3.2
x = np.arange(-2, 2, .2)
y = np.arange(-2, 2, .2)
X, Y = np.meshgrid(x, y)

# union
mur1=math.e**(-1*(X**2+Y**2));
mur2=1-math.e**(-1*(X**2+Y**2));

plt.rcParams['font.size'] = 14
plt.rcParams['font.family'] = 'Times New Roman'

# グラフの入れ物を用意する。
fig = plt.figure()
ax1 = fig.add_subplot(111, projection='3d')

# 軸のラベルを設定する。
ax1.set_xlabel('x')
ax1.set_ylabel('y')
ax1.set_zlabel('Union of R1 and R2')

# データプロットする。
ax1.plot_surface(X, Y, np.maximum(mur1,mur2), cmap='jet')

# グラフを表示する。
plt.show()
plt.close()


x = np.arange(-2, 2, .2)
y = np.arange(-2, 2, .2)
X, Y = np.meshgrid(x, y)

# union
mur1=math.e**(-1*(X**2+Y**2));
mur2=1-math.e**(-1*(X**2+Y**2));

plt.rcParams['font.size'] = 14
plt.rcParams['font.family'] = 'Times New Roman'

# グラフの入れ物を用意する。
fig = plt.figure()
ax1 = fig.add_subplot(111, projection='3d')

# 軸のラベルを設定する。
ax1.set_xlabel('x')
ax1.set_ylabel('y')
ax1.set_zlabel('Union of R1 and R2')

# データプロットする。
ax1.plot_surface(X, Y, np.minimum(mur1,mur2), cmap='jet')

# グラフを表示する。
plt.show()
plt.close()
```
